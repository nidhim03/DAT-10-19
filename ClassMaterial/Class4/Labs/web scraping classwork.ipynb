```python
import pandas as pd
import requests
import bs4              # this is the webscraper
```

```python
url='https://www.yelp.com/search?find_desc=Restaurants&find_loc=London%2C%20United%20Kingdom&ns=1&start=0'
req=requests.get(url)
```

```python
scraper=bs4.BeautifulSoup(req.text)
scraper
```

```python
type(scraper)
```

```python
class_selectors={'class':'link__09f24__1kwXV', 'class': 'link-color--inherit__09f24__3PYlA', 'class': 'link-size--inherit__09f24__2Uj95'}

page_links=scraper.find_all('a', class_selectors)
page_links
```

```python
page_links=[link.text for link in page_links]
link_texts=[name for name.tex in page_links if ".css" not in name]
link_texts
```
